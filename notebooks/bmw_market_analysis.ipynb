```python
df['Fuel_Type'].value_counts()
```

```python
df['is_hybrid']=(df['Fuel_Type']=='Hybrid').astype(int)
```

```python
df['is_petrol']=(df['Fuel_Type']=='Petrol').astype(int)
```

```python
df['is_diesel']=(df['Fuel_Type']=='Diesel').astype(int)
```

```python
df['is_electric']=(df['Fuel_Type']=='Electric').astype(int)
```

```python
pd.get_dummies(df['Fuel_Type'])
```

```python
df.columns.to_list()
```

```python
df['manual_transmission']=(df['Transmission']=='Manual').astype(int)
```

```python
df['automatic_transmission']=(df['Transmission']=='Automatic').astype(int)
```

```python
df=df.rename(columns={'Engine_Size_L':'Engine_Size'})
df
```

```python
df.columns.tolist()
```

```python
#make different columns for engine size
df['small_engine']=(df['Engine_Size']<=2).astype(int)
```

```python
df['medium_engine']=df['Engine_Size'].between(2,3).astype(int)
```

```python
df['Large_Engine']=(df['Engine_Size']>3).astype(int)
```

```python
df['Mileage_KM'].describe()
```

```python
def mileage_bucket(mileage):
    if mileage <= 60000:
        return 'Low Mileage'
    elif mileage <= 130000:
        return 'Medium Mileage'
    else:
        return 'High Mileage'

df['Mileage_Category'] = df['Mileage_KM'].apply(mileage_bucket)
```
